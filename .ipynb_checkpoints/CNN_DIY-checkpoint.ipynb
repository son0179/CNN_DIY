```python
cifar10_dir = 'datasets/cifar-10-batches-py'

# Cleaning up variables to prevent loading data multiple times (which may cause memory issue)
try:
   del X_train, y_train
   del X_test, y_test
   print('Clear previously loaded data.')
except:
   pass

X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

# As a sanity check, we print out the size of the training and test data.
print('Training data shape: ', X_train.shape)
print(y_train)
print('Training labels shape: ', y_train.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
```

```python
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = len(classes)
plt.axis('off')
samples_per_class = 10

for y, cls in enumerate(classes):
    idxs = np.flatnonzero(y_train == y) # y_train 에서 y라벨을 갖는 원소만 가져와 새 리스트 생성
    idxs = np.random.choice(idxs, samples_per_class, replace=False) # 7개만 랜덤으로 비복원추출(중복 X)
    for i, idx in enumerate(idxs):
        plt_idx = i * num_classes + y + 1
        plt.subplot(samples_per_class, num_classes, plt_idx) # (samples_per_class) * (num_classes) 개의 그래프중 plt_idx번째 작업
        plt.imshow(X_train[idx].astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)

plt.show()
```

```python
import numpy as np

print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
print(X_test.shape)
print(y_test.shape)

batch_size = 500
X_tra = np.vsplit(X_train,X_train.shape[0]//batch_size) 
X_tra = np.array(X_tra)
y_tra = np.split(y_train,y_train.shape[0]//batch_size) 
y_tra = np.array(y_tra)
X_tra=X_tra.astype(float)
print(X_tra.shape)
print(X_tra[0][0])
```

```python
#find good learning rate
import random
from layers import *
from update_grad import *

W1 = np.ones((9,3,16,16))
W2 = np.ones((15,9,16,16))
batch_size = 2000
epoch = 10

tmp = random.uniform(-3,-6)
learning_rate = 10**tmp 
conv_params = { "stride":1 , "pad":0}

conv1 = ConvLayer(W1.shape,conv_params,0,learning_rate)
relu1 = ReLULayer()
conv2 = ConvLayer(W2.shape,conv_params,0,learning_rate)
relu2 = ReLULayer()
affin = AffineLayer(learning_rate)
conv1_update = SgdM()
conv2_update = SgdM()
aff_update = SgdM()
for j in range(epoch):
    ans = 0
    loss_sum = 0
    
    for i in range(20):

        out = conv1.conv_forward(X_tra[i])
        #print("out : ", out[0,:10])

        out = relu1.ReLU_forward(out)
        #print("out : ", out[0,:10])

        out = conv2.conv_forward(out)
        #print("out : ", out[0,:10])

        out = relu2.ReLU_forward(out)
        #print("out : ", out[0,:10])

        out = affin.aff_forward(out)
        #print("out : ", out[0,:10])

        result= np.argmax(out ,axis = 1)
        ans += np.sum(result == y_tra[i])
        #print("out : ", out[0,:10])
        
        
        loss , dX = softmax(out,y_tra[i])
        loss_sum += loss

        dX , affdW = affin.aff_backward(out)
        dX = relu2.ReLU_backward(dX)
        dX , conv2dW = conv2.conv_backward(dX)
        dX = relu1.ReLU_backward(dX)
        dX , conv1dW = conv1.conv_backward(dX)
        
        conv1.w -= conv1.lr * conv1_update.update(conv1dW)
        conv2.w -= conv2.lr * conv2_update.update(conv2dW)
        affin.w -= affin.lr * aff_update.update(affdW)
        
    print("cor : ", ans / (X_tra.shape[0] * X_tra.shape[1]) , end = " // ")
    print("loss : ", loss_sum / X_tra.shape[0] , end = " // " )
    print("learning rate : ", learning_rate)
    
    
```
